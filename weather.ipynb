```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe =  OneHotEncoder()
```

```python
target = ohe.fit_transform(train[['weather-condition']])
```

```python
target = target.toarray()
print(target)
```

```python
from sklearn.linear_model import Ridge
```

```python
main_col = ["month","year","tempmax","tempmin","temp","dew","humidity","windgust","windspeed","winddir","sealevelpressure","cloudcover","visibility","solarradiation","solarenergy"]
```

```python
from sklearn.neighbors import LocalOutlierFactor


lof = LocalOutlierFactor()


outliers = lof.fit_predict(train[main_col])
```

```python
outliers
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='most_frequent') 


imputer.fit(train[main_col])


X_train_imputed = imputer.transform(train[main_col])
X_test_imputed = imputer.transform(test)
```

```python
rig = Ridge(alpha = .1)
```

```python
rig.fit(X_train_imputed, target)
```

```python
predictions = rig.predict(X_test_imputed)
```

```python
print(predictions)
```

```python
from numpy import argmax
```

```python
result = argmax(predictions)
```

```python
result = []
```

```python
for i in predictions:
    result.append(argmax(i))
```

```python
print(result)
```

```python

weather_pred = []
for i in result:
    if i == 0:
        weather_pred.append("clear-day")
    elif i == 1:
        weather_pred.append("partly-cloudy-day")
    else:
        weather_pred.append("rain")
```

```python
print(weather_pred)
```
